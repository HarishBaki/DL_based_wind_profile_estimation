```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# Latitude and longitude values
lats = [54.0143, 51.97, 56.44050, 51.706898, 51.6463, 
        51.9171, 51.9990, 52.306583, 53.71250, 54, 
        54.177256]
lons = [6.58385, 4.93, 8.15080, 3.034618, 2.95141, 
        3.6670, 3.2760, 4.008972, 7.1522, 6.623333, 
        7.891971]

# Site identifiers
site_ids = [f"cite{i+1}" for i in range(len(lats))]

# Initialize the plot
fig, ax = plt.subplots(figsize=(10, 10), subplot_kw={'projection': ccrs.PlateCarree()})

# Set extent to focus on Europe
ax.set_extent([0, 10, 50, 60], crs=ccrs.PlateCarree())

# Add geographic features
ax.add_feature(cfeature.LAND, color='lightgrey')
ax.add_feature(cfeature.OCEAN, color='lightblue')
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS, linestyle=':')

# Plot each site with its ID
for lat, lon, site_id in zip(lats, lons, site_ids):
    ax.plot(lon, lat, marker='o', color='red', markersize=5, transform=ccrs.PlateCarree())
    ax.text(lon + 0.3, lat, site_id, transform=ccrs.PlateCarree(), fontsize=8, color='black')

# Set title
plt.title("Site Locations on European Map")
plt.savefig("site_locations.png", dpi=300)
```
